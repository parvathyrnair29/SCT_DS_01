```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset (adjust path if needed)
file_path = "/content/API_SP.POP.TOTL_DS2_en_csv_v2_81108.csv"
df = pd.read_csv(file_path, skiprows=4)
# Age groups and population (in millions)
age_groups = ['0–14', '15–29', '30–44', '45–59', '60+']
populations = [370, 330, 270, 230, 210]

# Create histogram-like bar chart
plt.bar(age_groups, populations, color='grey', edgecolor='black')

# Add labels and title
plt.title("India's Population by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Population (in Millions)")

# Add population labels above bars
for i, value in enumerate(populations):
    plt.text(i, value + 5, f"{value} Mn", ha='center')

# Show chart
plt.tight_layout()
plt.show()
```
